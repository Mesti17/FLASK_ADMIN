```python
# %pip install git+https://github.com/JustAnotherArchivist/snscrape.git
import snscrape.modules.twitter as sntwitter
import pandas as pd 
```

```python
# %pip install textblob
from textblob import TextBlob
import re

def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ",tweet).split())

def analize_sentiment(tweet):
    analysis = TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0 :
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
    
def analize_susjectorobject(tweet):
    analysis = TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity == 0:
        return "netral"
    else:
        return "Negatif"
    
    
```

```python
tweet_list = []
```

```python
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('prabowo').get_items()):
    if i>100:
        break
    tweet_list.append([tweet.id, tweet.content, tweet.user.username, analize_sentiment(tweet.content)])
```

```python
tweet = pd.DataFrame(tweet_list, columns=['Tweet Id', 'Text', 'Username', 'Label'])

tweet.tail(20)
```
